# Function to get the data and save the dataframe to csv fil 

```python
class linkedin:
    def __init__(self):
        pass
    def data_file(job_name, url):    
        data = requests.get(url).text
        soup = BeautifulSoup(data)
        job_title = soup.findAll('h3', class_='base-search-card__title')
        company= soup.find_all('h4', class_="base-search-card__subtitle")
        link=soup.find_all('h4', class_="base-search-card__subtitle")
        loc= soup.find_all('span', class_='job-search-card__location')
        hire= soup.findAll('span', class_='result-benefits__text')
        img= soup.find_all('img')
        info=[job_title, company, loc,hire]
    
        trial=[]
        for i in info:
            temp=[]
            for el in i :
                z= el.text
                z=z.strip()
                print(z)
                temp.append(z) 
            my_list=temp
            trial.append(my_list)
        
        Title =trial[0]
        company = trial[1]
        location=trial[2]
        hire=trial[3]
        
        country=[]
        for i in location :
            x=i.split(",",3)
            country.append(x[-1])
      
        logo=[]
        for i in img:
            z=i['data-delayed-url']
            logo.append(z)
    
        links=[]
        for i in link:
            z=i.find("a")['href']
            links.append(z)
        
        
        dict={"Title":Title,"Company":company, "Location":location, "Country or State":country, "Company Logo":logo, "Company page Link":links}
    
        globals()[job_name]= pd.DataFrame(dict)
        job =pd.DataFrame(dict)
        job.to_csv('jobs.csv')
        return job
        
```
